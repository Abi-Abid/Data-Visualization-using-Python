### Line plots (Series/Dataframe)

##### What is a line plot? and why we use it?

A line chart or a line plot is a type of plot which displays information as a series of data points called 'markers' connected by straight line segments. It is a basic type of chart common in many feilds. Use line plot when we have a continous data set. These are best suited for trend-based visualizations of data over a period of time.

#### Let;s start with a case study

In 2010, Haiti suffered a catasptropic magnitude 7.0 earthquake. The quake caused wide spread devastation and loss of life and about 3 million people affested by this natural disater. As part of canadian humanitarian effort, The Government of Canada stepped up its effort in accepting refugees from Haiti. Let's quickly visualize this effort by using a line plot:

Question - Plot a line graph of immigration from Haiti using df.plot()

First, let's extract the data series of Haiti

```python
#since we have converted the years column into data type strings
#let's declare a variable that will allow us easily call upon the full range of year

years = list(map(str,range(1980,2014)))

#creating data sweries
haiti = df.loc['Haiti', years] #passing in years 1980-2013 to exclude the 'total' column
haiti.head()
```

```python
#let's plot a line plot by appending '.plot()' to 'haiti' dataframe
haiti.plot()
```

```python
#also, let;s label the x and y axis using 'ply.title()', 'plt.xlabel()', 'plt.ylabel()'
haiti.plot(kind = 'line')

plt.title('Immigration from Haiti')
plt.xlabel('Years')
plt.ylabel('No of Immigrants')

plt.show()
```

We can clearly notice how nuber of immigratants from haiti spiked up from 2010 as Canada stepped up its effoerts to accept refugees from Haiti. Lets annotate this spike in plot by using the 'plt.text()' method.

However, notice that years of type is 'string'. Let's change the type of index values to integer first.

```python
haiti.index = haiti.index.map(int)
haiti.plot(kind='line')

plt.title('Immigrants from Haiti')
plt.xlabel('Years')
plt.ylabel('No of Immigrants')

#annotate the 2010 earthquake
plt.text(2000,6000, '2010 Earthquake')
plt.show()

           
```

With just a few lines of code, we were able to quickly identify and visualize the spike in immigration!

Quick note on 'x' and 'y' values in 'plt.text(x, y, label)'

Since the x-axis(years) type is integer, we specified x as a year. The y-axis (no of immigrants), type is integer, so we can just specify the value y = 6000.

plt.text(2000, 6000, '2010 Earthquake')

if the years were storewd as type 'string', we would need to speify x as the index position of the year. eg., - 20th index is year 2000 since it is the 20th year with a base year of 1980.

plt.text(20, 6000, '2010 Earthquake') #year stored as string

We can easily add more countries to line plot to make meaningful comparisons for immigrations from different countries.

Question - Let's compare the no of immigrants from India and China from 1980 to 2013

```python
#first let's get the data set for China and India
df_CI = df.loc[['India', 'China'], years]
df_CI
```

```python
#let's plot line plot by passing in 'kind' parameter to 'plot()'
df_CI.plot(kind = 'line')
```

This doesn't look right

If we recall pandas plots tyhe indices on x-axis and the column as individual lines on the y-axis. Since df_CI is a dataframe with the 'country' s the index and 'years' as the columns, we must first transpose the dataframe using 'transpose()' method to swap the row and columns.

```python
df_CI = df_CI.transpose()
df_CI.head()
```

```python
df_CI.index = df_CI.index.map(int)  #changes the index values of df_CI to type integer for plotting
df_CI.plot(kind = 'line')

plt.title('Immigrants from China and India')
plt.ylabel('No of Immigrants')
plt.xlabel('Years')

plt.show()
```

From the above plot we can observe that the China and India have very similar trends through theyears.

Note - How come we didn't need to transpose Haiti's df before plottong like we did for df_CI

Line plot is a handy tool to display various dependent variables againstone independent variable. However, it is recommende that no more than 5 to 10 lines on a single grapgh, any more than that it will be difficult top interpret.

```python
#let's compare the trend of top 5 countries that contributed the most to Canada's Immigration
df.sort_values('Total', axis = 0, ascending = False, inplace = True)   # sorts 'Total' column according to top 5 countries
df_top5 = df.head(5)
df_top5 = df_top5[years].transpose()

df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='line', figsize = (14,8))

plt.title('Immigration trend of top 5 Countyries')
plt.xlabel('Number of Immigrants')
plt.ylabel('years')

plt.show()
```
