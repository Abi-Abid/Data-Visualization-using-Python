# Plotting directly with Matplotlib

### Objectives

- Create and customize basic plots directly with Matplotlib on dataset

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv'

df = pd.read_csv(url)
df.head()
```

```python
df.set_index('Country', inplace = True)
df.head()
```

```python
years = list(map(str, range(1980 , 2014)))
```

### Line Plot

A line plot displays the relationship between two continuous variables over a continuous interval, showing the trend or pattern of the data.

Let's create a line plot to visualize the immigrants (to Canada) trend during 1980 to 2013.
We need the Total of year-wise immigrants,

We will create a new dataframe for only columns containing the years
then, we will apply sum() on the dataframe

We can do create a line plot directly on axes by calling plot function plot()

```python
y = list(map(str, years))

df_line = df[years]

total_immigrants = df_line.sum()
total_immigrants
```

```python
fig, ax = plt.subplots()

ax.plot(total_immigrants)

ax.set_title('Immigrants between 1980 to 2013')
ax.set_xlabel('Years')
ax.set_ylabel('Total Immigrants')

plt.show()
```

The plot function populated the x-axis with the index values (years), and the y-axis with the column values (population).
However, notice how the years were not displayed because they are of type string.

Therefore, let's change the type of the index values to integer for plotting.

```python
fig ,ax = plt.subplots()
total_immigrants.index = total_immigrants.index.map(int)
ax.plot(total_immigrants)
ax.set_title('Immigrants Between 1980 to 2013 to Canada')
ax.set_xlabel('Years')
ax.set_ylabel('No of Immigrants')
plt.show()
```

##### Let's customize the above graph appearance

```python
fig, ax = plt.subplots()
total_immigrants.index = total_immigrants.index.map(int)
ax.plot(total_immigrants, marker = 's', markersize = 5, color = 'green', linestyle = 'dotted')

ax.set_title('Immigrants to Canada from 1980 to 2013')
ax.set_xlabel('Years')
ax.set_ylabel('No of Immigrants')

plt.show()
```

```python
#let's try to include the background grid, a legend and try to change the limits on the axis
fig, ax = plt.subplots()

total_immigrants.index = total_immigrants.index.map(int)

ax.plot(total_immigrants, marker = 's', markersize = 5, color = 'r', linestyle = 'dotted')

ax.set_title('Immigrants to Canada from 1980 to 2013')
ax.set_xlabel('years')
ax.set_ylabel('No of Immigrants')

plt.xlim(1975, 2015)
plt.grid(True)
plt.legend(['Immigrants'])
plt.show()
```

## Scatter plot

A scatter plot visualizes the relationship between two continuous variables, displaying individual data points as dots on a two-dimensional plane, allowing for the examination of patterns, clusters, and correlations.

Let's created a Scatter plot to visualize the immigrants (to Canada) trend during 1980 to 2013.
We need the Total of year-wise immigrants,

We will create a new dataframe for only columns containing the years
then, we will apply sum() on the dataframe

You can do create a scatter plot directly on ax by calling plot function scatter()

```python
fig, ax = plt.subplots(figsize = (8,4))

ax.scatter(years, total_immigrants, marker = 'o', s = 20, color = 'darkblue')

ax.set_title('Immigrants between 1980 to 2013')
ax.set_xlabel('Years')
ax.set_ylabel('No of Immigrants')

plt.grid(True)
ax.legend(['Immirants'], loc = 'upper center')

plt.show()
         
```

## Bar Plot

A bar plot represents categorical data with rectangular bars, where the height of each bar corresponds to the value of a specific category, making it suitable for comparing values across different categories.

Let's create a bar plot to visualize the top 5 countries that contribued the most immigrants to Canada from 1980 to 2013.

Apply sort_values function on the 'Total' column of our data

We will create a new dataframe for only columns containing the years
then, we will apply sum() on the dataframe and can create a separatedataframe for top five countries

You can further use the names of the countries to label each bar on the plot

```python
df.sort_values('Total', ascending = False, axis = 0, inplace = True)

df_top5 = df.head()
df_bar_5 = df_top5.reset_index()

label = list(df_bar_5.Country)
label
```

```python
#the third name is too long to fit on the x-axis as label. Let's fix this using indexing.
label[2] = 'UK'
label
```

## Histogram

A histogram is a way of representing the frequency distribution of numeric dataset. The way it works is it partitions the x-axis into bins, assigns each data point in our dataset to a bin, and then counts the number of data points that have been assigned to each bin. So the y-axis is the frequency or the number of data points in each bin. Note that we can change the bin size and usually one needs to tweak it so that the distribution is displayed nicely.

Let's find out the frequency distribution of the number (population) of new immigrants from the various countries to Canada in 2013?



```python
df_country = df.groupby(['Country'])['2013'].sum().reset_index()

fig, ax = plt.subplots(figsize=(10,4))

ax.hist(df_country['2013'])
ax.set_title('New Immigrants in 2013')
ax.set_xlabel('No of Immigrants')
ax.set_ylabel('No of Countries')

plt.show()
```

Our plot doesnot match with the bars
By default, the histrogram method breaks up the dataset into 10 bins. The figure below summarizes the bin ranges and the frequency distribution of immigration in 2013
The hist function retuns list of arrays with 1. counts and 2. bins. we can fetch that using unpacking functionality
and further use the bins as x-ticks

```python
fig, ax = plt.subplots(figsize = (10, 40))
counts = ax.hist(df_country['2013'])
ax.set_title('New Immigrants in 2013')
ax.set_xlabel('No of Immigrants')
ax.set_ylabel('No of Countries')
ax.set_xticks(list(map(int, counts[1])))
ax.legend(['Immigrants'])
plt.show()
                       
```

## Pie Chart

A pie chart represents the proportion or percentage distribution of different categories in a dataset using sectors of a circular pie.
Let's create a pie chart representing the 'Total Immigrants' for the year 1980 to 1985

```python
fig, ax = plt.subplots()

ax.pie(total_immigrants[0:5], labels = years [0:5], colors = ['gold', 'blue', 'lightgreen', 'coral', 'cyan'], autopct = '%1.1f%%', explode = [0,0,0,0,0.1])
ax.set_aspect('equal')
plt.title('Distribution of Immigrants from 1980 to 1985')
plt.show()
```

## Subplotting

Let us explore how to display more than one plot on the same figure and specify the number of rows and columns to be created to the subplots function.

For instance, letâ€™s create a line and scatter plot in one row
plt.subplots()
You can use the same functions using which you plotte lne and scatter plots at the start of this lab
Both the subplots will be sharing the same y-axis as the data in the y-axis is the same. So, assign the â€˜Shareyâ€™ parameter as True in the code below. Also notice the use of 'suptitle'



```python
fig, axs = plt.subplots(1,2, sharey = True)

axs[0].plot(total_immigrants)
axs[0].set_title('Line plot on Immigrants')

axs[1].scatter(total_immigrants.index, total_immigrants)
axs[1].set_title('Scatter plot on Immigrants')

axs[0].set_ylabel('Number of Imigrants')

fig.suptitle('Subplotting example', fontsize = 15)
fig.tight_layout()
plt.show()
```

You can also implement the subplotting with add_subplot() as below:-

```python
fig = plt.figure(figsize = (8,4))

axs1 = fig.add_subplot(1,2,1)
axs1.plot(total_immigrants)
axs1.set_title('Line plot on Immigrants')

ax2 = fig.add_subplot(1,2,2)
ax2.barh(total_immigrants.index, total_immigrants)
ax2.set_title('Bar plot on immigrants')

fig.suptitle('Subplotting Example', fontsize = 15)
fig.tight_layout()
plt.show()
```
